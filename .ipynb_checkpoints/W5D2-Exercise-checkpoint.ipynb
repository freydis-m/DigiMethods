> **Ex. 5** Load the HTML you downloaded in Ex. 3. For each post, extract price, square meter size and "Ejerudgift". You should create three different lists that contain each variable across posts.

```python
source = BeautifulSoup(requests.get("https://www.boliga.dk/resultat?propertyType=3&zipCodes=2200&page=1").content, "html.parser")
import re

housinglist = []

for line in source.findAll("app-housing-list-item"):
    
    price = line.find("div", class_="price m-0 d-flex align-items-center").text
    _sqmeter = line.findAll("div", class_="detail")
    sqmeter = _sqmeter[1].text
    
    
    #ejerudgift
    link = line.a.get("href")
    followlink = BeautifulSoup(requests.get(f"https://www.boliga.dk{link}").content, "html.parser")
    finddetails = followlink.findAll("span", class_="d-md-none my-auto")
    ejerudgift = finddetails[6].text
    
    a=[price, sqmeter, ejerudgift]
    print(a)
    
    housinglist.append(a)
```

```python
clearedlist = []
    
for element in housinglist:
    price = element[0]
    sqmeters = element[1]
    ejerudgift = element[2]
    
    
    price1= re.sub("[^0-9]", "", price)
    sqmeters1 = re.sub("[^0-9]", "", sqmeters)
    ejerudgift1 = re.sub("[^0-9]", "", ejerudgift)

    
    cleared = [price1, sqmeters1, ejerudgift1]
    clearedlist.append(cleared)
        
print(clearedlist)
```

> **Ex. 6:** Make a scatter plot of square meter size vs. extracted price. Then make a new variable that 
measures price per square meter and scatter plot this against "Ejerudgift". Can you say anything about how "Ejerudgift" influences square meter price?

```python
#not much, since the plot doesn't seem to be sorted.....
```

```python
import matplotlib.pyplot as plt


sqmeter = [item[1] for item in clearedlist]
price = [item[0] for item in clearedlist]
ejerudgift = [item[2] for item in clearedlist]

plt.figure(figsize=(10, 7), dpi= 120)

plt.title("Comparison of housing prices and sizes")
plt.xlabel("Price in DKK")
plt.ylabel("Size in square meters")

plt.xticks(range(0,len(price)), price, rotation="90")
plt.yticks(range(0,len(sqmeter)), sqmeter)

plt.plot(price, sqmeter, "o")
```

```python
priceprsq =[]
for element in clearedlist:
    new = int(element[0]) / int(element[1])
    priceprsq.append(int(new))
    
print(priceprsq)

plt.figure(figsize=(10, 7), dpi= 120)

plt.title("Comparison of price per square meter and ejerudgift")
plt.xlabel("Price per square meter in DKK")
plt.ylabel("Ejerudgift in DKK")


plt.plot(priceprsq, ejerudgift, "o")
```

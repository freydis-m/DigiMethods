**Exercise 5.1.1.1** Read in the dataset "nchs_death_causes.csv" as a list of lists. You will need to use the `csv` package and then will want to go through each row appending it to a list you create.

```python
import csv

with open("nchs_death_causes.csv", "r") as file: 
    csv_reader = csv.reader(file, delimiter=',')
    nchsdata = []
    
    for row in csv_reader:
        print(row)
        nchsdata.append(row)
    
```

**Exercise 5.1.1.2** Make a list that is a subset of the data, containing only one cause of death and one state.

```python



cause_of_death = "Alzheimer's disease"
state = "Washington"


filteredlist = []



for element in nchsdata:
    if element[2]==cause_of_death and element[3]==state:
        filteredlist.insert(0, element)
        
        
    
print(filteredlist)


    

"""
#alzheimersdeaths_alabama = [[1999, 772, 17.8],
#                            [2000, 895, 20.4],
#                            [2001, 1103, 25.1],
#                            [2002, 1189, 27.0],
 #                           [2003, 1268, 28.8],
  #                          [2004, 1385, 31.4],
   #                         [2005, 1501, 33.4],
    #                        [2006, 1497, 32.7],
     #                       [2007, 1517, 32.6],
      #                      [2008, 1518, 32.1],
       #                     [2009, 1521, 31.6],
        #                    [2010, 1523, 31.2],
         #                   [2011, 1486, 29.7],
          #                  [2012, 1390, 27.2],
           #                 [2013, 1398, 26.7],
            #                [2014, 1885, 35.3],
             #               [2015, 2282, 41.8],
              #              [2016, 2507, 45]]
"""
```

**Exercise 6.1.1.1** Read in the dataset "nchs_death_causes.csv" using `pandas` methods.

```python
import pandas as pd 

nchs_deaths = pd.read_csv("nchs_death_causes.csv")

nchs_deaths
```

**Exercise 6.1.1.2** Make a `DataFrame` that is a subset of the data, containing only one cause of death and one state.

```python
injuries = nchs_deaths[nchs_deaths["Cause Name"].isin(["Unintentional injuries"])]
injuries_alabama = injuries[injuries["State"].isin(["Alabama"])]
```

```python
injuries_alabama
```

**Exercise 6.1.1.3** Plot a line graph that shows the development of the death rate over time. Pay attention to whether your plot is intuitive.

```python
import matplotlib.pyplot as plt

injuries_alabama_series = injuries_alabama["Age-adjusted Death Rate"]
injuries_alabama_series.index = injuries_alabama["Year"]


plt.plot(injuries_alabama_series)
```

**Exercise 6.1.1.4** Make your figure self-contained and export it/save it to your machine.

```python
plt.figure(figsize=(7, 5), dpi= 120)
plt.plot(injuries_alabama_series)

plt.title("Injury death rate, Alabama 1999-2018")
plt.xlabel("Year")
plt.ylabel("Age-adjusted death rate")
plt.xticks(range(1999,1999+len(injuries_alabama_series)), injuries_alabama["Year"].tolist().reverse(), rotation=90)
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
chan_paranormal_p1 = BeautifulSoup(requests.get("http://boards.4channel.org/x/").content, "html.parser")
```

```python
page1=[]

for line in chan_paranormal_p1.findAll("div", class_="thread"):
    threads = {}
    
    threads["postnum"]=len(page1)
    threads["Original post"]=line.find("blockquote", class_="postMessage").text
    threads["OP date"]=line.find("span", class_="dateTime").text
    threads["OP image"]=line.find("a", class_="fileThumb").get("href")
    
    for reply in line.findAll("div", class_="postContainer replyContainer"):
        replies=[]
        replies.append(reply.find("blockquote", class_="postMessage").text)
    
    threads["OP replies"]=replies
    page1.append(threads)
    
print(page1)
```
